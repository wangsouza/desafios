```
# Basic setup
import numpy as np
import matplotlib.pyplot as plt
```

A função de ativação **sigmoid** é uma função muito popular para redes neurais. A entrada para a função é transformada em um valor entre 0,0 e 1,0. As entradas acima de 1,0 são transformadas no valor de 1,0, da mesma forma os valores menores que 0,0 são ajustadas para 0,0.

A função **tangente hiperbólica (tanh)** é uma função de ativação não linear de formato muito semelhante a função sigmoid, porém produz valores entre -1,0 e 1,0.

O grande problema entre as duas funções é devido ao fato que elas saturam, ou seja, valores muito pequenos se ajustam a 0 ou -1, respectivamente a função sigmoid e tanh, da mesma forma com valores muito grandes se ajustam a 1. Essa sensibilidade limitada ocasiona um problema chamado de **gradiente de fuga**.
O gradiente de fuga ocorre geralmente em redes neurais com camadas mais profundas, esse erro é propagado de volta pela rede e usado para atualizar os pesos.

A função **RELU** retorna 0 para todos os valores negativos, mas aumenta lienarmente para os valores positivos. A função RELU não é saturável, ou seja, não ocorre gradiente de fuga, o que torna uma função computacionalmente leve e empiricamente superior à função sigmoid e tanh na maioria das aplicações.

```
# Plot the sigmoid function
z = np.linspace(-5, 5, 200)

def sigmoid(z):
  return 1.0 / (1 + np.exp(-z))

def tanh(z):
  # return np.tanh(z)
  return (np.exp(z) - np.exp(-z))/(np.exp(z) + np.exp(-z))

def relu(z):
  return np.maximum(0, z)

# Plot activations
plt.figure(figsize=(9,7))
plt.plot(z, sigmoid(z), "b-", linewidth=2, label='Sigmoid', alpha=0.8)
plt.plot(z, tanh(z), "r-", linewidth=2, label='tanh', alpha=0.8)
plt.plot(z, relu(z), "g-", linewidth=2, label='ReLU',alpha=0.8)

# Format plot
plt.plot([-5, 5], [1, 1], 'k--')
plt.plot([-5, 5], [0, 0], 'k--')
plt.plot([-5, 5], [-1, -1], 'k--')
plt.axis([-4, 4, -1.2, 1.2])
plt.legend()
plt.grid(True)
plt.title("Activation Function", fontsize=14)
plt.xlabel('WX+b')
plt.ylabel('Activation Output')
plt.show()
```
